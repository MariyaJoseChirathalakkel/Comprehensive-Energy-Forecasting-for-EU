## Get  information of the missing values in uk_data

```python
# Get column names and data types
column_info = uk_data.dtypes.reset_index()
column_info.columns = ['Column_Name', 'Data_Type']

# Count missing values for each column
missing_values_count = uk_data.isnull().sum().reset_index()
missing_values_count.columns = ['Column_Name', 'Missing_Values']

# Merge column names, data types, and missing values count
column_info = pd.merge(column_info, missing_values_count, on='Column_Name')

# Save the information to a CSV file
column_info.to_csv('column_info.csv', index=False)
```

## Data Cleaning

### Step 1: Remove the first 3505 rows


```python
# Remove the first 3505 rows
uk_data_trimmed = uk_data.iloc[3505:].reset_index(drop=True)

# Verify the first few rows of the trimmed DataFrame
print(uk_data_trimmed.head())

```

### Step 2: Fill the missing values by considering the average of the preceeding and suceeding rows 

```python

def interpolate_missing_values_inplace(df, columns):
    """
    Fill missing values in the specified columns of the DataFrame using linear interpolation in place.

    Parameters:
    df (pd.DataFrame): The DataFrame containing the data.
    columns (list): The list of column names to interpolate.
    """
    # Ensure the columns exist in the DataFrame
    for column in columns:
        if column not in df.columns:
            raise ValueError(f"Column '{column}' not found in DataFrame")

    # Perform linear interpolation for the specified columns in place
    for column in columns:
        df[column].interpolate(method='linear', inplace=True)



# Specify the columns to interpolate
columns_to_interpolate = ['GB_GBN_load_actual_entsoe_transparency',
'GB_GBN_load_forecast_entsoe_transparency',
'GB_GBN_price_day_ahead',
'GB_GBN_solar_capacity',
'GB_GBN_solar_generation_actual',
'GB_GBN_solar_profile',
'GB_GBN_wind_capacity',
'GB_GBN_wind_generation_actual',
'GB_GBN_wind_profile',
'GB_GBN_wind_offshore_capacity',
'GB_GBN_wind_offshore_generation_actual',
'GB_GBN_wind_offshore_profile',
'GB_GBN_wind_onshore_capacity',
'GB_GBN_wind_onshore_generation_actual',
'GB_GBN_wind_onshore_profile',
'GB_NIR_load_actual_entsoe_transparency',
'GB_NIR_load_forecast_entsoe_transparency',
'GB_NIR_solar_capacity',
'GB_NIR_wind_onshore_capacity',
'GB_NIR_wind_onshore_generation_actual',
'GB_UKM_load_actual_entsoe_transparency',
'GB_UKM_load_forecast_entsoe_transparency',
'GB_UKM_solar_capacity',
'GB_UKM_solar_generation_actual',
'GB_UKM_wind_capacity',
'GB_UKM_wind_generation_actual',
'GB_UKM_wind_offshore_capacity',
'GB_UKM_wind_offshore_generation_actual',
'GB_UKM_wind_onshore_capacity',
'GB_UKM_wind_onshore_generation_actual']

# Interpolate the specified columns in place
interpolate_missing_values_inplace(uk_data_trimmed, columns_to_interpolate)

# Verify the first few rows of the DataFrame after interpolation
print(uk_data_trimmed.head())
```

### Step 3: Convert 'utc_timestamp' column to datetime format


```python
uk_data_trimmed['utc_timestamp'] = pd.to_datetime(uk_data_trimmed['utc_timestamp'])
```
